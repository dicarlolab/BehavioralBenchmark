```python
inds = []
positions = []
for i in range(0, 255, 2):    
    for j in range(0, 255, 2):
        positions.append((i,j))
for i, m in enumerate(dataset.meta):
    if m['var'] in ['V3', 'V6'] and m['obj'] in hvm.OBJECT_SET_KPROTOCOL:
        inds.append(i)
```

```python
pixel_masks = dataset.pixel_masks[inds,:,:]
results = {}

for inds, filename in zip([dataset.IT_NEURONS, dataset.V4_NEURONS],
                          ['fine_IT_bootstrap_figure_ground_results_mcc', 'fine_V4_bootstrap_figure_ground_results_mcc']):
    selection = np.ix_(np.ravel(inds), np.ravel(dataset.IT_NEURONS), [0,1,2,3,4])
    F = dataset.multiunit_time_bins[selection]
    results = {}
    for bootstrap_ind in range(20):
        rng = np.random.RandomState(bootstrap_ind)
        bstrapinds = rng.choice(F.shape[1], F.shape[1], replace=True)
        f = F[:, bstrapinds, :]
        results[bstrapind] = Parallel(verbose=700, n_jobs=15)(delayed(get_figure_ground)(f, position) for position in positions)
    
    pk.dump(results, open(filename, 'wb'))
```
